7. I have a character with this specification:
defense_overall 6.0<br>
offense_overall 7.0<br>
skill_effect_overall 4.0<br>
difficulty_overall 3.0<br>
movement_spd 250<br>
magic_defense 9<br>
mana 380<br>
hp_regen 37<br>
physical_atk 119<br>
physical_defense 24<br>
hp 2689<br>
attack_speed 0.75<br>
mana_regen 16.0<br>

Please predict its role using DecisionTree classification algorithm <br>

```python
#7
# import library
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score

# Encode the 'role' column
label_encoder = LabelEncoder()
df_excel['role'] = label_encoder.fit_transform(df_excel['role'])

# separate features and target variable
features = df_excel[['defense_overall','offense_overall', 'skill_effect_overall', 
                     'difficulty_overall', 'movement_spd', 'magic_defense', 
                     'mana', 'hp_regen', 'physical_atk', 'physical_defense', 
                     'hp', 'attack_speed', 'mana_regen']]
target = df_excel['role']  

# split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# create a Decision Tree Classifier
clf = DecisionTreeClassifier(random_state=42)

# train the model
clf.fit(X_train, y_train)

# make predictions on the test set
y_pred = clf.predict(X_test)

# evaluate the model (only accuracy)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:}")

# predict the role for the given character
character_features = pd.DataFrame({
    'defense_overall': [6.0],
    'offense_overall': [7.0],
    'skill_effect_overall': [4.0],
    'difficulty_overall': [3.0],
    'movement_spd': [250],
    'magic_defense': [9],
    'mana': [380],
    'hp_regen': [37],
    'physical_atk': [119],
    'physical_defense': [24],
    'hp': [2689],
    'attack_speed': [0.75],
    'mana_regen': [16.0]
})

# predict the role for the given character
predicted_role_encoded = clf.predict(character_features)
predicted_role = label_encoder.inverse_transform(predicted_role_encoded)

# get the predicted role
print(f"The predicted role for the character is: {predicted_role[0]}")
```
